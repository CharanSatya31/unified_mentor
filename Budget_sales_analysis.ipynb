```python
import pandas as pd

adventureworks_df = pd.read_excel('/content/AdventureWorks_Database.xlsx', sheet_name=None)
budget_df = pd.read_excel('/content/Budget.xlsx', sheet_name=None)

adventureworks_sheet_names = adventureworks_df.keys()
budget_sheet_names = budget_df.keys()

adventureworks_preview = {sheet: adventureworks_df[sheet].head() for sheet in adventureworks_sheet_names}
budget_preview = {sheet: budget_df[sheet].head() for sheet in budget_sheet_names}

print(adventureworks_sheet_names)
print(adventureworks_preview)
print(budget_sheet_names)
print(budget_preview)
```

```python
from datetime import datetime

customers_df = adventureworks_df['Customers']

current_year = datetime.now().year
customers_df['Age'] = current_year - customers_df['BirthDate'].dt.year

relevant_columns = ['Gender', 'MaritalStatus', 'YearlyIncome', 'TotalChildren', 'Education', 'Occupation', 'HouseOwnerFlag', 'NumberCarsOwned', 'CommuteDistance', 'Age']
customers_analysis_df = customers_df[relevant_columns]

customers_analysis_df.head()
```

```python
# Display basic statistics of the selected columns
customers_analysis_df.describe(include='all')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.countplot(data=customers_analysis_df, x='Gender')
plt.title('Gender Distribution')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=customers_analysis_df, x='MaritalStatus')
plt.title('Marital Status Distribution')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data=customers_analysis_df, x='YearlyIncome', kde=True)
plt.title('Yearly Income Distribution')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=customers_analysis_df, x='Education')
plt.title('Education Level Distribution')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=customers_analysis_df, x='Occupation')
plt.title('Occupation Distribution')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=customers_analysis_df, x='HouseOwnerFlag')
plt.title('House Ownership')
plt.xticks([0, 1], ['Non-owner', 'Owner'])
plt.show()


plt.figure(figsize=(10, 6))
sns.countplot(data=customers_analysis_df, x='NumberCarsOwned')
plt.title('Number of Cars Owned')
plt.show()


plt.figure(figsize=(10, 6))
sns.countplot(data=customers_analysis_df, x='CommuteDistance')
plt.title('Commute Distance Distribution')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data=customers_analysis_df, x='Age', kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt

gender_counts = customers_analysis_df['Gender'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.show()

marital_counts = customers_analysis_df['MaritalStatus'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(marital_counts, labels=marital_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Marital Status Distribution')
plt.show()

education_counts = customers_analysis_df['Education'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(education_counts, labels=education_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Education Level Distribution')
plt.show()

occupation_counts = customers_analysis_df['Occupation'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(occupation_counts, labels=occupation_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Occupation Distribution')
plt.show()
```

```python
import seaborn as sns

numerical_vars = customers_analysis_df[['Age', 'YearlyIncome', 'TotalChildren', 'NumberCarsOwned']]
sns.pairplot(numerical_vars)
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numerical_vars = customers_analysis_df[['Age', 'YearlyIncome', 'TotalChildren', 'NumberCarsOwned']]

# Calculate the correlation matrix
correlation_matrix = numerical_vars.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
budget_clean = budget_data.iloc[3:].reset_index(drop=True)
budget_clean.columns = budget_data.iloc[2]
budget_clean = budget_clean.dropna(how='all').reset_index(drop=True)

for col in budget_clean.columns[3:]:
    budget_clean[col] = pd.to_numeric(budget_clean[col], errors='coerce')

budget_clean['Total'] = budget_clean.iloc[:, 3:].sum(axis=1)

plt.figure(figsize=(10, 6))
sns.barplot(data=budget_clean, x='Category', y='Total', ci=None)
plt.title('Total Sales by Category')
plt.xticks(rotation=45)
plt.show()
```
